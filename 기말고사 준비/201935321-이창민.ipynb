### 11. 

```python
import pandas as pd
import numpy as np
#csv 파일 불러오기
df = pd.read_csv("fish_length_data.csv")
#무작위로 추출된 표본 1000개
X = np.random.choice(df['length'], 1000, replace=False)


# (a)
print(f"X의 표본평균: {round(np.mean(X), 4)}")
# (b)
print(f"X의 표본표준편차: {round(np.std(X, ddof=1), 4)}")
```

```python
import matplotlib.pyplot as plt

df = pd.read_csv("fish_length_data.csv")

w = 10
h = 6
fig = plt.figure(figsize=(w, h))
ax1 = fig.add_subplot(2, 2, 1)
ax2 = fig.add_subplot(2, 2, 2)
ax3 = fig.add_subplot(2, 2, 3)
ax4 = fig.add_subplot(2, 2, 4)

ax1.hist(df['length'], bins=10, edgecolor='w')
ax1.vlines(sum(df['length'])/10, 0, 1, label='mean', colors='red')  #기대값:평균
ax2.hist(df['length'], bins=100, edgecolor='w') 
ax1.vlines(sum(df['length'])/100, 0, 1, label='mean', colors='red')  #기대값:평균
ax3.hist(df['length'], bins=1000, edgecolor='w')
ax1.vlines(sum(df['length'])/1000, 0, 1, label='mean', colors='red')  #기대값:평균
ax4.hist(df['length'], bins=len(df['length']), edgecolor='w')
ax1.vlines(np.mean(df['length']), 0, 1, label='mean', colors='red')  #기대값:평균
plt.show()
```

### 12. 

```python
S, MEAN, STD = 2600000, 145, 5
data =  MEAN + np.random.randn(S) * STD

prob = len([ i for i in data if i >=145 and i<150]) / S
print(f"무작위로 추출한 학생의 키가 145이상 150미만일 확률: {round(prob, 4)}")
```

```python
from scipy.stats import norm    # 정규분포관련

data.sort()
X = data
Y = norm.pdf(X, MEAN, STD)   #확률밀도함수(pdf)에 대한 표준정규분포(norm)

plt.plot(X, Y)
plt.hlines(0, min(X), max(X), alpha=0.3)
plt.fill_between(X, Y, where=(X>=145) & (X<150), color='red', alpha=0.5)
plt.show()
```

```python
import scipy.stats as sp
t = 32
p = 0.1


rv = sp.binom(t, p)

xx = np.arange(t + 1)
plt.bar(xx, rv.pmf(xx), align='center')
plt.grid()
plt.show()


```
